Exercise: Find the recent and the earliest travelled date and mean distance travelled for each start city

```python
df1 = df.iloc[:-1, :]
df1.tail()
df1["START_DATE*"] = pd.to_datetime(df1["START_DATE*"])
df1.dtypes
```

```python
res = df1.groupby("START*").agg({"MILES*": "mean", "START_DATE*":["min", "max"]}).head()
# res
res.reset_index(inplace=True)
res.columns= ["Start_City","Avg_Dist","Earliest_Trip", "Recent_Trip"]
res
```

```python
# tranfer into a csv file
res.to_csv("response.csv")
```

```python
store_sales = pd.read_csv("store_sales.csv")
store_sales.head()
```

```python
# find out the mean in the month of january
store_sales["Jan"].agg("mean")
```

```python
# find out the mean of all months
# store_sales[:].agg("mean")

# Alternate solution
store_sales.apply("mean")
```

```python
# adding a column depend on jan sale which are eligible or not eligible for bonus
import numpy as np
def bonus_func(sales_col):
    elig_values = np.where(sales_col > 10, 'eligible', 'not eligible')
    return elig_values
store_sales["Jan_bonus"] = bonus_func(store_sales["Jan"])
store_sales
```

```python
newdf = store_sales.select_dtypes(include=["int64", "float64"]).apply(bonus_func)
newdf
```

```python
newdf2 = pd.concat([store_sales, newdf], axis = 1)
newdf2
```
